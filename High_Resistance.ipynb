```python
import numpy as np
import matplotlib.pyplot as plt
```

Formula:

$R = \frac{t}{C.ln(\frac{d_1}{d_2})}$

Inputs

```python
# All length units are in mm and time unit in seconds

t = np.array([2, 4, 6, 8, 10]) # Time required to discharge

C = 1e-6 # Capacitor Value 1μF

# For R1 = 10MΩ

d1 = 60 # First throw on the Galvanometer before Discharge

d2 = np.array([50, 45, 40, 35, 30]) # First throw on the Galvanometer after Discharge


# For R2 = 20MΩ

D1 = 150 # First throw on the Galvanometer before Discharge

D2 = np.array([135, 120, 115, 100, 90]) # First throw on the Galvanometer after Discharge
```

Graph

```python
# Graph to Find the "High Resistance"

plt.plot(t, np.log(d1/d2), color = 'green', marker = 'o')  # Plot for R1
plt.plot(t, np.log(D1/D2), color = 'red', marker = 'o')  # Plot for R2
plt.xlabel("Time t (s)")
plt.ylabel("ln(deflect1/deflect2)")
plt.grid(True)
plt.show()

slope1, intercept = np.polyfit(t, np.log(d1/d2), 1)
print("Slope for R1 =",np.round(1/slope1, decimals=3))

print('-------------------------')

slope2, intercept = np.polyfit(t, np.log(D1/D2), 1)
print("Slope for R2 =",np.round(1/slope2, decimals=3))
```

Calculations

```python
Slope1 = t / np.log(d1/d2)

R1 = np.mean(Slope1)/C

print('t/ln(d1/d2) =',Slope1.round(3))
print('-------------------------')
print('t/ln(d1/d2)_mean =',np.mean(Slope1.round(4)))
print('-------------------------')
print('Experimental Value for R1 =',"{:.3e}".format(R1), 'Ω')
print('-------------------------')

Slope2 = t / np.log(D1/D2)

R2 = np.mean(Slope2)/C

print('t/ln(D1/D2) =',Slope2.round(3))
print('-------------------------')
print('t/ln(D1/D2)_mean =',np.mean(Slope2.round(4)))
print('-------------------------')
print('Experimental Value for R2 =',"{:.3e}".format(R2), 'Ω')
print('-------------------------')

R1_G = 1/(slope1*C)
print('Graphical Value for R1 =',"{:.3e}".format(R1_G), 'Ω')

R2_G = 1/(slope2*C)
print('-------------------------')
print('Graphical Value for R2 =',"{:.3e}".format(R2_G), 'Ω')
```
